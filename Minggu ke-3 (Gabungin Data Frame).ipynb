Data Integration

```python
import pandas as pd
import numpy as np
```

```python
#Marging
biodata=pd.DataFrame({'id':['id0','id1','id2','id3'],
                    'JK':['P','P','L','L'],
                    'Usia':['19','21','18','23']})
biodata
```

```python
nilai=pd.DataFrame({'id':['id0','id1','id2','id3'],
                     'IPK':['3,94','3,34','3,75','3,11'],
                     'Nilai':['A','AB','AB','B']})
nilai
```

cara ngegabungin dua data frame

```python
result1=pd.merge(biodata,nilai,on='id')
result1
```

Merubah urutan colum

```python
result1=result1[['id','JK','Usia','Nilai','IPK']]
result1
```

kalo dia berdasarkan dua var

```python
#multiple key
biodata1=pd.DataFrame({'id':['id0','id1','id2','id3'],
                        'Kelas':['kalkulus','ADE','kalkulus','ADW'],
                        'JK':['P','P','L','L'],
                        'Usia':['19','21','18','23']})
biodata1
```

```python
nilai1=pd.DataFrame({'id':['id0','id1','id2','id3'],
                     'Kelas':['kalkulus','ADK','MO','ADW'],
                     'IPK':['3,94','3,34','3,75','3,11'],
                     'Nilai':['A','AB','AB','B']})
nilai1
```

```python
result2=pd.merge(biodata1,nilai1,on=['id','Kelas'])
result2
```

Merge methods (menggabung dua data frame) (default how="inner)

```python
# Merge how="inner"
result_inner=pd.merge(biodata1,nilai1,how='inner',on=['id','Kelas'])
result_inner
```

```python
# Merge how="left"
result_left=pd.merge(biodata1,nilai1,how='left',on=['id','Kelas'])
result_left
```

```python
# Merge how="Right"
result_right=pd.merge(biodata1,nilai1,how='right',on=['id','Kelas'])
result_right
```

```python
# Merge how="Outer"
result_outer=pd.merge(biodata1,nilai1,how='outer',on=['id','Kelas'])
result_outer
```

Selain metode Merge ada juga metode Join untuk menggabung dua Data Frame

```python
# Join
biodata3=pd.DataFrame({'JK':['P','P','L'],
                       'Usia':['19','21','18']},
                       index=['id0','id1','id2'])
biodata3
```

```python
nilai3=pd.DataFrame({'Kelas':['Kalkulus','ADK','MO'],
                     'IPK':['3,94','3,34','3,75']},
                     index=['id0','id2','id3'])
nilai3
```

Data Frame biodata ditambah Data Frame nilai Base on Data Frame biodata

```python
result3=biodata3.join(nilai3) 
result3
```

Data Frame nilai ditambah Data Frame biodata Base on Data Frame nilai

```python
result4=nilai3.join(biodata3) 
result4
```

Join juga ada 2 base -> outer,inner

```python
# Outer -> Seluruh Data Masuk
result5=nilai3.join(biodata3,how='outer')
result5
```

```python
# Inner -> Hanya data yang lengkap yang masuk
result6=nilai3.join(biodata3,how='inner')
result6
```

```python
result7=pd.merge(biodata3, nilai3, left_index=True, right_index=True, how='outer')
result7
```

```python
result8=pd.merge(biodata3, nilai3, left_index=True, right_index=True, how='inner')
result8
```

```python
biodata4 = pd.DataFrame({'JK': ['P', 'L', 'P', 'L'], 
                      'Usia': ['19', '21', '18', '23'],
                   'key': ['id0', 'id1', 'id0', 'id1']})
nilai4 = pd.DataFrame({'Nilai': ['A', 'B'],
                       'IPK': ['3.44', '3.74']},
                      index=['id0', 'id1'])
result9 = biodata4.join(nilai4, on='key')
result9
```

```python
## Concatenating
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})
frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

```python
# Add category
result = pd.concat(frames, keys=['X', 'Y', 'Z'])
result
```

```python
# Default join=outer (join all rows)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

#axis=1 is for concated by y-axis; default axis=0
result = pd.concat([df1, df4], axis=0) 
result
```

```python
# Join = inner (join only rows that exist in both datasets) 
# hanya yang ada di df1 dan df4
result = pd.concat([df1, df4], axis=1, join='inner')
result
```
