# Asociation Rule

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data=pd.read_csv('D:\Praktikum\dataminingc_week6.csv')
data
```

```python
data=data.drop('Kode',axis=1)
data.head()
```

```python
frequent_itemsets = apriori(data, min_support=0.3, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence",min_threshold=0.7)
rules
```

```python
rules.sort_values(by=['support'],ascending=False)
```

```python
rules.sort_values(by=['confidence'],ascending=False)
```

```python
rules.sort_values(by=['lift'],ascending=False)
```

```python
rules.sort_values(by=['confidence','lift'],ascending=False)
```

```python
rules.sort_values(by=['support','confidence','lift'],ascending=False)
```
